The name of the variables must follow these requirements:

- Use only alphanumeric characters and 'underscore' ( _ )
- DO NOT contain blank spaces
- DO NOT start with a number
- In Python the name of variables is *case sensitive*, i.e.: `Python` != `python`
- There are reserved words that cannot be used to name variables, for example: `False`, `True`, `None`, `for`, `return`

Find the complete list of reserved keywords for python 3.8 __[here](https://docs.python.org/3.8/reference/lexical_analysis.html#keywords)__

## <font color=#FF1F58> >>> Numeric Types</font>


Numeric types in Python are objects that store numeric values. They are created by assigning a value to them.

These are the most common numeric values:

- `int`: integers (`1`, `234789`)

- `float`: decimals (`0.3`, `723.36`)

You can perform arithmetic operation in Python using these operators:
- addition: `3 + 4`
- subtraction: `2 - 0.9`
- multiplication: `2.1 * 300` 
- division: `2 / 9`
- exponentiation: `10**2` 
- modulus: `10 % 8` (`2`in the example, returns the remaining numbers by dividing the first number from the second)

## <font color=FF1F58> >>> Strings</font>

Strings are sequences of characters within quotes. In Python, string variables can be declared using either single quotes (`''`), double quotes (`""`) or triple quotes `(''' ''')`.

```python
name = 'Carlos'
last_name = "Smith"
favourite_quotation = '''That's one small step for a man, a giant leap for mankind'''
print('Name: ', name)
print('Last name: '+ last_name)
print('Favourite quotation: ', favourite_quotation)
```

There are characters with specific meanings: `\t`, `\n`.... If you want those characters to be printed, use raw strings, prefixing the string with an `r`

```python
string_with_newline = 'hello\nthere'
print(string_with_newline)
print('---')
string_without_newline = r'hello\nhere'
print(string_without_newline)
```

##  <font color=#FF1F58>3. Type Casting</font>

#### Convert to integer: `int()`

```python
float_to_int = int(5.6)
print(float_to_int)

bool_true_to_int = int(True)
print(bool_true_to_int)

bool_false_to_int = int(False)
print(bool_false_to_int)
```

```python
string_to_int = int('Hello there!')
```

#### Convert to float:`float()`

```python
int_to_float = float(23)
print(int_to_float)

bool_true_to_float = float(True)
print(bool_true_to_float)
      
bool_false_to_int = float(False)
print(bool_false_to_int)
```

```python
string_to_float = float('Hello there!')
```

#### Convert to string: `str()`

```python
int_to_string = str(2)
print(int_to_string)

float_to_string = str(3.14159)
print(float_to_string)

bool_true_to_string = str(True)
print(bool_true_to_string)

bool_false_to_string = str(False)
print(bool_false_to_string)
```

#### Convert to boolean: `bool()`

```python
positive_integer_to_bool = bool(5)
negative_integer_to_bool = bool(-2)
float_number_to_bool = bool(0.34)
zero_to_bool = bool(0)


print('5 --> ', positive_integer_to_bool)
print('-2 -->', negative_integer_to_bool)
print('0.34 -->'  , float_number_to_bool)
print('0 -->', zero_to_bool)
```

```python
empty_string_to_bool = bool('')
hello_to_bool = bool('Hello there!')
true_string_to_bool = bool("True")
false_string_to_bool = bool("False")

print('empty string --> ', empty_string_to_bool)
print('hello -->', hello_to_bool)
print('"True" string -->'  , true_string_to_bool)
print('"False" string -->', false_string_to_bool)
```
